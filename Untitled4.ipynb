```python


import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM, GRU, SimpleRNN
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error
import math
```

# Téléchargement et chargement des données


```python

import requests
import io

url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/airline-passengers.csv"
df = pd.read_csv(url)
df.head()
```

#Visualisation des données

```python
plt.figure(figsize=(12, 6))
plt.plot(df['Passengers'])
plt.title('Evolution du nombre de passagers aériens (1949-1960)')
plt.xlabel('Mois')
plt.ylabel('Nombre de passagers (milliers)')
plt.grid(True)
plt.show()
```

#Prétraitement des données

```python
# Extraction de la colonne Passengers
dataset = df['Passengers'].values.astype('float32')
dataset = dataset.reshape(-1, 1)

# Normalisation Min-Max
scaler = MinMaxScaler(feature_range=(0, 1))
dataset = scaler.fit_transform(dataset)

# Division train/test (75%/25%)
train_size = int(len(dataset) * 0.75)
test_size = len(dataset) - train_size
train = dataset[0:train_size, :]
test = dataset[train_size:len(dataset), :]

print(f"Taille dataset: {len(dataset)}")
print(f"Taille train: {len(train)}")
print(f"Taille test: {len(test)}")
```

# Fonction de création des séquences

```python
def getdata(data, lookback):
    X, Y = [], []
    for i in range(len(data) - lookback - 1):
        X.append(data[i:(i + lookback), 0])
        Y.append(data[i + lookback, 0])
    return np.array(X), np.array(Y)

lookback = 1
X_train, y_train = getdata(train, lookback)
X_test, y_test = getdata(test, lookback)

# Reshape pour Keras [samples, timesteps, features]
X_train = X_train.reshape(X_train.shape[0], X_train.shape[1], 1)
X_test = X_test.reshape(X_test.shape[0], X_test.shape[1], 1)

print(f"Forme X_train: {X_train.shape}")
print(f"Forme y_train: {y_train.shape}")
```

# Partie 1: Modèle RNN Simple

```python
# Construction du modèle RNN
model_rnn = Sequential()
model_rnn.add(SimpleRNN(5, input_shape=(lookback, 1)))
model_rnn.add(Dense(1))
model_rnn.compile(loss='mean_squared_error', optimizer='adam')
model_rnn.summary()

# Entraînement
history_rnn = model_rnn.fit(X_train, y_train, epochs=50, batch_size=1, verbose=1)

# Prédiction
y_pred_rnn = model_rnn.predict(X_test)

# Transformation inverse pour obtenir les valeurs originales
y_test_orig = scaler.inverse_transform(y_test.reshape(-1, 1))
y_pred_rnn_orig = scaler.inverse_transform(y_pred_rnn)

# Calcul de l'erreur
mse_rnn = mean_squared_error(y_test_orig, y_pred_rnn_orig)
print(f"Erreur quadratique moyenne (RNN): {mse_rnn:.2f}")

# Visualisation
plt.figure(figsize=(14, 6))
plt.plot(y_test_orig, label='Valeurs réelles', linewidth=2)
plt.plot(y_pred_rnn_orig, label='Valeurs prédites (RNN)', linewidth=2)
plt.title('Comparaison des valeurs réelles et prédites - RNN Simple')
plt.ylabel('Nombre de passagers')
plt.legend()
plt.show()
```

# Partie 2: Comparaison avec LSTM et **GRU**

## 2a. Modèle LSTM

```python
# Construction du modèle LSTM
model_lstm = Sequential()
model_lstm.add(LSTM(5, input_shape=(lookback, 1)))
model_lstm.add(Dense(1))
model_lstm.compile(loss='mean_squared_error', optimizer='adam')
model_lstm.summary()

# Entraînement
history_lstm = model_lstm.fit(X_train, y_train, epochs=50, batch_size=1, verbose=1)

# Prédiction
y_pred_lstm = model_lstm.predict(X_test)
y_pred_lstm_orig = scaler.inverse_transform(y_pred_lstm)

# Calcul de l'erreur
mse_lstm = mean_squared_error(y_test_orig, y_pred_lstm_orig)
print(f"Erreur quadratique moyenne (LSTM): {mse_lstm:.2f}")
```

# 2b. Modèle GRU

```python
# Construction du modèle GRU
model_gru = Sequential()
model_gru.add(GRU(5, input_shape=(lookback, 1)))
model_gru.add(Dense(1))
model_gru.compile(loss='mean_squared_error', optimizer='adam')
model_gru.summary()

# Entraînement
history_gru = model_gru.fit(X_train, y_train, epochs=50, batch_size=1, verbose=1)

# Prédiction
y_pred_gru = model_gru.predict(X_test)
y_pred_gru_orig = scaler.inverse_transform(y_pred_gru)

# Calcul de l'erreur
mse_gru = mean_squared_error(y_test_orig, y_pred_gru_orig)
print(f"Erreur quadratique moyenne (GRU): {mse_gru:.2f}")
```


# 2c. Comparaison des résultats

```python
# Visualisation comparative
plt.figure(figsize=(16, 8))
plt.plot(y_test_orig, label='Valeurs réelles', linewidth=3, color='black')
plt.plot(y_pred_rnn_orig, label=f'RNN (MSE: {mse_rnn:.2f})', alpha=0.8)
plt.plot(y_pred_lstm_orig, label=f'LSTM (MSE: {mse_lstm:.2f})', alpha=0.8)
plt.plot(y_pred_gru_orig, label=f'GRU (MSE: {mse_gru:.2f})', alpha=0.8)
plt.title('Comparaison des modèles RNN, LSTM et GRU (lookback=1)')
plt.ylabel('Nombre de passagers')
plt.legend()
plt.show()

# Tableau comparatif
comparison = pd.DataFrame({
    'Modèle': ['RNN', 'LSTM', 'GRU'],
    'MSE': [mse_rnn, mse_lstm, mse_gru]
})
print(comparison)
```

# Partie 3: Augmentation du lookback (lookback=2)

```python
# Recréation CORRECTE des données avec lookback=2
lookback2 = 2

X_train2, y_train2 = getdata(train, lookback2)
X_test2, y_test2 = getdata(test, lookback2)

X_train2 = X_train2.reshape(X_train2.shape[0], X_train2.shape[1], 1)
X_test2 = X_test2.reshape(X_test2.shape[0], X_test2.shape[1], 1)

print(f"Formes avec lookback=2:")
print(f"X_train2: {X_train2.shape}, y_train2: {y_train2.shape}")
print(f"X_test2: {X_test2.shape}, y_test2: {y_test2.shape}")

# IMPORTANT: Recréer y_test_orig avec la bonne taille
y_test2_orig = scaler.inverse_transform(y_test2.reshape(-1, 1))
print(f"y_test2_orig: {y_test2_orig.shape}")
```

# 3a. RNN avec lookback=2

```python
# Construction du modèle
model_rnn2 = Sequential()
model_rnn2.add(SimpleRNN(5, input_shape=(lookback2, 1)))
model_rnn2.add(Dense(1))
model_rnn2.compile(loss='mean_squared_error', optimizer='adam')

print("Entraînement du RNN avec lookback=2...")
history_rnn2 = model_rnn2.fit(X_train2, y_train2, epochs=50, batch_size=1, verbose=0)

# Prédiction
y_pred_rnn2 = model_rnn2.predict(X_test2)
y_pred_rnn2_orig = scaler.inverse_transform(y_pred_rnn2)

# Calcul MSE avec les bonnes dimensions
mse_rnn2 = mean_squared_error(y_test2_orig, y_pred_rnn2_orig)
print(f"Erreur quadratique moyenne RNN (lookback=2): {mse_rnn2:.2f}")
```

# 3b. LSTM avec lookback=2

```python
model_lstm2 = Sequential()
model_lstm2.add(LSTM(5, input_shape=(lookback2, 1)))
model_lstm2.add(Dense(1))
model_lstm2.compile(loss='mean_squared_error', optimizer='adam')

print("Entraînement du LSTM avec lookback=2...")
history_lstm2 = model_lstm2.fit(X_train2, y_train2, epochs=50, batch_size=1, verbose=0)

y_pred_lstm2 = model_lstm2.predict(X_test2)
y_pred_lstm2_orig = scaler.inverse_transform(y_pred_lstm2)
mse_lstm2 = mean_squared_error(y_test2_orig, y_pred_lstm2_orig)
print(f"Erreur quadratique moyenne LSTM (lookback=2): {mse_lstm2:.2f}")
```

# 3c. GRU avec lookback=2

```python
model_gru2 = Sequential()
model_gru2.add(GRU(5, input_shape=(lookback2, 1)))
model_gru2.add(Dense(1))
model_gru2.compile(loss='mean_squared_error', optimizer='adam')

print("Entraînement du GRU avec lookback=2...")
history_gru2 = model_gru2.fit(X_train2, y_train2, epochs=50, batch_size=1, verbose=0)

y_pred_gru2 = model_gru2.predict(X_test2)
y_pred_gru2_orig = scaler.inverse_transform(y_pred_gru2)
mse_gru2 = mean_squared_error(y_test2_orig, y_pred_gru2_orig)
print(f"Erreur quadratique moyenne GRU (lookback=2): {mse_gru2:.2f}")
```

# 3d. Comparaison lookback=1 vs lookback=2

```python
# Réaffichage des résultats avec lookback=1 pour comparaison
print("RAPPEL - Résultats avec lookback=1:")
print(f"RNN: {mse_rnn:.2f}, LSTM: {mse_lstm:.2f}, GRU: {mse_gru:.2f}")

print("\nRésultats avec lookback=2:")
print(f"RNN: {mse_rnn2:.2f}, LSTM: {mse_lstm2:.2f}, GRU: {mse_gru2:.2f}")

# DataFrame comparatif
results_comparison = pd.DataFrame({
    'Modèle': ['RNN (lookback=1)', 'LSTM (lookback=1)', 'GRU (lookback=1)',
               'RNN (lookback=2)', 'LSTM (lookback=2)', 'GRU (lookback=2)'],
    'MSE': [mse_rnn, mse_lstm, mse_gru, mse_rnn2, mse_lstm2, mse_gru2]
})

print("\nComparaison des performances:")
print(results_comparison)

# Visualisation comparative
plt.figure(figsize=(16, 10))

# Lookback=1
plt.subplot(2, 1, 1)
plt.plot(y_test_orig, label='Réelles', linewidth=2, color='black')
plt.plot(y_pred_rnn_orig, label=f'RNN (MSE: {mse_rnn:.2f})')
plt.plot(y_pred_lstm_orig, label=f'LSTM (MSE: {mse_lstm:.2f})')
plt.plot(y_pred_gru_orig, label=f'GRU (MSE: {mse_gru:.2f})')
plt.title('Modèles avec lookback=1')
plt.ylabel('Nombre de passagers')
plt.legend()

# Lookback=2
plt.subplot(2, 1, 2)
plt.plot(y_test2_orig, label='Réelles', linewidth=2, color='black')
plt.plot(y_pred_rnn2_orig, label=f'RNN (MSE: {mse_rnn2:.2f})')
plt.plot(y_pred_lstm2_orig, label=f'LSTM (MSE: {mse_lstm2:.2f})')
plt.plot(y_pred_gru2_orig, label=f'GRU (MSE: {mse_gru2:.2f})')
plt.title('Modèles avec lookback=2')
plt.ylabel('Nombre de passagers')
plt.legend()

plt.tight_layout()
plt.show()
```

# Partie 4 : Ajout d'une deuxième couche récurrente

# 4a. RNN avec 2 couches

```python
model_rnn_2layers = Sequential()
model_rnn_2layers.add(SimpleRNN(5, return_sequences=True, input_shape=(lookback2, 1)))
model_rnn_2layers.add(SimpleRNN(5))
model_rnn_2layers.add(Dense(1))
model_rnn_2layers.compile(loss='mean_squared_error', optimizer='adam')

print("Entraînement du RNN à 2 couches...")
history_rnn_2l = model_rnn_2layers.fit(X_train2, y_train2, epochs=50, batch_size=1, verbose=0)

y_pred_rnn_2l = model_rnn_2layers.predict(X_test2)
y_pred_rnn_2l_orig = scaler.inverse_transform(y_pred_rnn_2l)
mse_rnn_2l = mean_squared_error(y_test2_orig, y_pred_rnn_2l_orig)
print(f"Erreur quadratique moyenne RNN 2 couches: {mse_rnn_2l:.2f}")
```

# 4b. LSTM avec 2 couches

```python
model_lstm_2layers = Sequential()
model_lstm_2layers.add(LSTM(5, return_sequences=True, input_shape=(lookback2, 1)))
model_lstm_2layers.add(LSTM(5))
model_lstm_2layers.add(Dense(1))
model_lstm_2layers.compile(loss='mean_squared_error', optimizer='adam')

print("Entraînement du LSTM à 2 couches...")
history_lstm_2l = model_lstm_2layers.fit(X_train2, y_train2, epochs=50, batch_size=1, verbose=0)

y_pred_lstm_2l = model_lstm_2layers.predict(X_test2)
y_pred_lstm_2l_orig = scaler.inverse_transform(y_pred_lstm_2l)
mse_lstm_2l = mean_squared_error(y_test2_orig, y_pred_lstm_2l_orig)
print(f"Erreur quadratique moyenne LSTM 2 couches: {mse_lstm_2l:.2f}")
```

# 4c. GRU avec 2 couches

```python
model_gru_2layers = Sequential()
model_gru_2layers.add(GRU(5, return_sequences=True, input_shape=(lookback2, 1)))
model_gru_2layers.add(GRU(5))
model_gru_2layers.add(Dense(1))
model_gru_2layers.compile(loss='mean_squared_error', optimizer='adam')

print("Entraînement du GRU à 2 couches...")
history_gru_2l = model_gru_2layers.fit(X_train2, y_train2, epochs=50, batch_size=1, verbose=0)

y_pred_gru_2l = model_gru_2layers.predict(X_test2)
y_pred_gru_2l_orig = scaler.inverse_transform(y_pred_gru_2l)
mse_gru_2l = mean_squared_error(y_test2_orig, y_pred_gru_2l_orig)
print(f"Erreur quadratique moyenne GRU 2 couches: {mse_gru_2l:.2f}")
```

# 4d. Comparaison finale

```python
# Résultats finaux
final_comparison = pd.DataFrame({
    'Configuration': [
        'RNN 1 couche', 'LSTM 1 couche', 'GRU 1 couche',
        'RNN 2 couches', 'LSTM 2 couches', 'GRU 2 couches'
    ],
    'MSE': [mse_rnn2, mse_lstm2, mse_gru2, mse_rnn_2l, mse_lstm_2l, mse_gru_2l]
})

print("COMPARAISON FINALE (lookback=2):")
print(final_comparison)

# Visualisation finale
plt.figure(figsize=(16, 12))

# 1 couche
plt.subplot(2, 1, 1)
plt.plot(y_test2_orig, label='Valeurs réelles', linewidth=3, color='black')
plt.plot(y_pred_rnn2_orig, label=f'RNN (MSE: {mse_rnn2:.2f})')
plt.plot(y_pred_lstm2_orig, label=f'LSTM (MSE: {mse_lstm2:.2f})')
plt.plot(y_pred_gru2_orig, label=f'GRU (MSE: {mse_gru2:.2f})')
plt.title('Modèles avec 1 couche récurrente (lookback=2)')
plt.ylabel('Nombre de passagers')
plt.legend()

# 2 couches
plt.subplot(2, 1, 2)
plt.plot(y_test2_orig, label='Valeurs réelles', linewidth=3, color='black')
plt.plot(y_pred_rnn_2l_orig, label=f'RNN 2 couches (MSE: {mse_rnn_2l:.2f})')
plt.plot(y_pred_lstm_2l_orig, label=f'LSTM 2 couches (MSE: {mse_lstm_2l:.2f})')
plt.plot(y_pred_gru_2l_orig, label=f'GRU 2 couches (MSE: {mse_gru_2l:.2f})')
plt.title('Modèles avec 2 couches récurrentes (lookback=2)')
plt.ylabel('Nombre de passagers')
plt.legend()

plt.tight_layout()
plt.show()
```
